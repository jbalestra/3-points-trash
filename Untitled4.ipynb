```python
import pandas as pd #Importando a biblioteca pandas, e chamando ela de pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Mall_Customers.csv') #Lendo o aquivo com os dados

count_male = 0;
count_female = 0;

for i in df['Gender']:
  if (i == 'Male'):
    count_male += 1
  else:
    count_female += 1

print('A porcentagem de sexos de pessoas que frequentam shoppings Ã© de: ')
print(df['Gender'].value_counts() / df.shape[0] * 100  )
plt.show()


labels = 'Male', 'Female'
sizes = [count_male, count_female]
fig, ax = plt.subplots()
ax.pie(sizes, labels = labels , autopct='%1.1f%%')
plt.show()


print('A mÃ©dia da idade das pessoas que frequentam shoppings Ã© de: ',round (df['Age'].mean()),('anos')) # df['Age'].mean(), uma funÃ§Ã£o do pandas para fazer a mÃ©dia da coluna Age

plt.hist(df["Age"])
plt.show()

print('A mÃ©dia de salario anual (k$) das pessoas que frequentam shoppings Ã© de: ',round (df['Annual Income (k$)'].mean()),('k$'))

plt.hist(df['Annual Income (k$)'])
plt.show()



```
